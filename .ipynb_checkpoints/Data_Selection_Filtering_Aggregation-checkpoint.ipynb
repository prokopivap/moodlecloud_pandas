```python
import pandas as pd
```

```python
df = pd.read_csv('AirBnB_NY/cleaned_airbnb_data.csv', parse_dates=['last_review'])
```

```python
df.loc[:, ('name', 'price', 'price_category', 'length_of_stay_category')]
```

```python
#df
df.iloc[0:10, [1, 9, 16, 17]]
```

```python
filtering_list = ['Manhattan', 'Brooklyn']
df[df['neighbourhood_group'].isin(filtering_list)]
```

```python
#Further filter the dataset to include only listings with a price greater than $100 and a
#number_of_reviews greater than 10.
df[(df['price'] > 100) & (df['number_of_reviews'] > 10)] 
```

```python
#Select columns of interest such as neighbourhood_group, price, minimum_nights, number_of_reviews, price_category 
#and availability_365 for further analysis.
df1 = df[['neighbourhood_group', 'price', 'minimum_nights', 'number_of_reviews', 'price_category', 'availability_365']]
df1
```

```python
group_df = df.groupby(['neighbourhood_group', 'price_category'])
group_df[['price', 'minimum_nights']].mean()
```

```python
#group_df[['number_of_reviews', 'availability_365']].mean()
df1.groupby(['neighbourhood_group', 'price_category']).agg({'number_of_reviews': ['mean'], 'availability_365': ['mean']})
```

```python
#Sort the data by price in descending order and by number_of_reviews in ascending order.
df1.sort_values(['price', 'number_of_reviews'], ascending=[False, True])
```

```python
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# Create a ranking of neighborhoods     based on the total number of listings       and the average price.

# Group by neighbourhood_group to calculate the total number of listings and average price
grouped = df1.groupby('neighbourhood_group').agg(
    total_listings=('price', 'size'),     # Total number of listings
    average_price=('price', 'mean')       # Average price
).reset_index()

print(grouped)





# Rank based on total listings (higher is better)
grouped['listings_rank'] = grouped['total_listings'].rank(ascending=False)

# Rank based on average price (higher is better)
grouped['price_rank'] = grouped['average_price'].rank(ascending=False)

# Combine the ranks by summing them
grouped['combined_rank'] = grouped['listings_rank'] + grouped['price_rank']

# Sort by the combined rank to get the final ranking
final_ranking = grouped.sort_values('combined_rank')

print(final_ranking)
```
